```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.random.randint(1,12,5)
y=np.random.randint(1,10,5)
```

```python
x
```

```python
y
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(x,y**2,color='green',linewidth=4)
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(x,y**2,color='green',lw=4,linestyle='-.',marker='D',markersize=10,markerfacecolor='red',markeredgecolor='blue',markeredgewidth=5)
```

```python
figure=plt.figure()
axes=figure.add_axes([0,0,1,1])
axes.plot(x,y**2,color='green',lw=4,linestyle='-.',marker='D',markersize=10,markerfacecolor='red',markeredgecolor='blue',markeredgewidth=5)
axes.set_xlim(0,20)
axes.set_ylim(0,100)
```
