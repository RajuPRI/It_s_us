```python
ls_x=np.linspace(-4,4,50)
```

```python
ls_y=ls_x **2
```

```python
plt.plot(ls_x,ls_y)
plt.show()
```

```python
plt.plot(ls_x,ls_y,'.')
plt.show()
```

```python
plt.plot(ls_x,ls_y,'r.')
plt.show()
```

```python
plt.plot(ls_x,np.sin(ls_x),'r.')
plt.plot(ls_x,np.cos(ls_x),'g--')
plt.show()
```
