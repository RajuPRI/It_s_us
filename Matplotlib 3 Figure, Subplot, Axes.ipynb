```python
x1=np.arange(1,9)
x1
```

```python
y1=np.arange(3,27,3)
```

```python
plt.plot(x1,y1)
```

```python
##error
plt.subplot(2,2,1)
plt.plot(x1,y1,'red')
plt.subplot(2,2,2)
plt.plot(y1,y1,'blue')
plt.subplot(2,2,3)
plt.plot(x1,x1,'yellow')
plt.subplot(2,2,4)
plt.plot(y1,x1,'black')
```

```python
plt.subplot(2,2,1)
plt.plot(x1,x1*x1,'red')
plt.subplot(2,2,2)
plt.plot(x1,np.sqrt(x1),'blue')
plt.subplot(2,2,3)
plt.plot(x1,np.exp(x1),'yellow')
plt.subplot(2,2,4)
plt.plot(y1,np.log10(x1),'black')
```
